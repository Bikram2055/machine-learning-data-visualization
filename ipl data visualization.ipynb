## visualization

```python

winners = pd.DataFrame({'winner': data['winner']})
count_wins = winners.value_counts()
labels = [x[0] for x in count_wins.keys()]

fig, ax = plt.subplots(figsize=(20,12))
ax = plt.pie(x= count_wins, autopct='%.1f%%', labels= labels)
plt.title('MOST WINS')
plt.legend()
plt.show()
```

```python
plt.bar(labels, count_wins)
plt.xticks(rotation = 90)
plt.title('most wins')
plt.xlabel('teams -->')
plt.ylabel('no. of wins -->')
plt.show()
```

## most game played in city

## toss decision

```python
teams = data['toss_winner'].unique()
decision_making = pd.DataFrame([], columns=['Toss winner', 'Decision', 'Times'])
for id, element in enumerate(teams):
    team_bat = data[(data['toss_winner']==element) & (data['toss_decision']=='bat')]  # if toss winner choose bat that is team bat
    team_field = data[(data['toss_winner']==element) & (data['toss_decision']=='field')]  # if toss winner choose field that is team field
    
    # append in decision making
    decision_making = decision_making.append({'Toss winner': element,
                           'Decision': 'bat', 'Times': team_bat['toss_winner'].count()}, ignore_index=True)
    decision_making = decision_making.append({'Toss winner': element,
                           'Decision': 'field', 'Times': team_field['toss_winner'].count()}, ignore_index=True)

sns.catplot(x="Toss winner", y="Times", hue="Decision", data=decision_making, kind='bar', height=5, aspect=2)
plt.xticks(rotation= 90)
plt.title('Toss decision of teams')
plt.xlabel('ipl teams')
plt.ylabel('toss decision')
plt.show()
```
